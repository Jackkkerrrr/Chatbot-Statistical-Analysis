```python
# remove repetitive information
def remove_repetitive_information(sentence_list):
  """return a sentence list with the removed repetitive information where the similarty score is greater than 0.9"""
  i = 0
  while i < len(sentence_list):
    vectorizer = CountVectorizer()
    cm = vectorizer.fit_transform(sentence_list)
    similarity_scores = cosine_similarity(cm[i], cm)
    similarity_scores = similarity_scores.flatten()

    index_list = []
    for j in range(len(similarity_scores)):
      if similarity_scores[j] >= 1:
        index_list.append(j)

    if i in index_list:
      index_list.remove(i)
    sentence_list = [item for index, item in enumerate(sentence_list) if index not in index_list]
    i += 1
  return sentence_list
sentence_list = remove_repetitive_information(sentence_list)

len(sentence_list)
```

```python
# remove questions
def remove_questions(sentence_list):
  """return a sentence list with all the question sentences removed"""
  q_word = ['what','why','when','where','how','who','?']
  for sentence in sentence_list:
    for q in q_word:
      if q in sentence.lower():
        sentence_list.remove(sentence)
        break
  return sentence_list
sentence_list = remove_questions(sentence_list)
len(sentence_list)
```

```python
# A function to return a random greeting response to a users greeting
def greeting_response(text):
  text = text.lower()

  # Bots greeting response
  bot_greetings = ['howdy', 'hi', 'hey', 'hello','hola']

  # users greeting
  user_greetings = ['hi','hey','hello','wassup','greetings','hola']

  for word in text.split():
    if word in user_greetings:
      return random.choice(bot_greetings)
```

```python
def index_sort(list_var):
  length = len(list_var)
  list_index = list(range(0, length))

  x = list_var
  for i in range(length):
    for j in range(length):
      if x[list_index[i]] > x[list_index[j]]:
        #swap
        temp = list_index[i]
        list_index[i] = list_index[j]
        list_index[j] = temp

  return list_index
```

```python
#Create the bots response
def bot_response(user_input):
  user_input = user_input.lower()

  sentence_list.append(user_input)
  bot_response = ''
  vectorizer = CountVectorizer()

  cm = vectorizer.fit_transform(sentence_list)

  similarity_scores = cosine_similarity(cm[-1], cm)

  similarity_scores_list = similarity_scores.flatten()
  index = index_sort(similarity_scores_list)
  index = index[1:]
  response_flag = 0
  j = 0
  for i in range(len(index)):
    if similarity_scores_list[index[i]] > 0.0:
      bot_response = bot_response + ' ' + sentence_list[index[i]]
      response_flag = 1
      j += 1
    if j > 2:
      break
  if response_flag == 0:
    bot_response = bot_response + ' ' + "I apologize, I don't understand"
  sentence_list.remove(user_input)
  return bot_response
```

```python
#Start the chat
print('Doc Bot: I am Doctor Bot or Doc bot for short, I will answer your queries about high blood pressure. If you want to exit, type bye.')

exit_list = ['exit', 'see you later', 'bye','quit', 'break']
while(True):
  user_input = input()
  #print(type(user_input))
  if user_input.lower() in exit_list:
    print('Doc Bot: Chat with you later')
    break
  else:
    if greeting_response(user_input) != None:
      print('Doc Bot:' + greeting_response(user_input))
    else:
      print('Doc Bot:' + bot_response(user_input))
```

Categorizes those questions

```python
def categorize_questions(questions):
  """return a dictionary containing the categorized questions, where the key is each category"""
  question_copy = questions.copy()

  i = 0
  question_categories = {}
  while len(question_copy) != 0:
    question_categories[i] = []
    vectorizer = CountVectorizer()
    cm = vectorizer.fit_transform(question_copy)
    similarity_scores = cosine_similarity(cm[0], cm)
    similarity_scores_list = similarity_scores.flatten()
    for j in range(len(similarity_scores_list)):
      if similarity_scores_list[j] >= 0.2:
        question_categories[i].append(question_copy[j])
    for q in question_categories[i]:
      question_copy.remove(q)
    i += 1
  return question_categories
```

```python
def get_question_category(questions, categories):
  """Return a list showing the type of questions with the same order as the question"""
  result = []
  for question in questions:
    for q_type in categories:
      if question in categories[q_type]:
        result.append(q_type)
        break
  for i in range(len(result)):
    if result[i] == 0:
      result[i] = 'General Information about High Blood Pressure'
    if result[i] == 1:
      result[i] = 'Understanding Blood Pressure Numbers'
    if result[i] == 2:
      result[i] = 'Concerns about Hypertension and Its Effects:'
    if result[i] == 3:
      result[i] = 'Support and Communication with Medical Teams'
    if result[i] == 4:
      result[i] = 'How to be a supportive partner to the medical team.'
  return result
```
